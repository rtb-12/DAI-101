```python
identifier = []
cat = []
num = []
date_time = []

for column in df.columns:
    if column in ['Country', 'Tech Sector']: 
        identifier.append(column)
    elif pd.api.types.is_numeric_dtype(df[column]):
        if df[column].nunique() == df.shape[0]:
            identifier.append(column)
        elif df[column].nunique() <= 10:
            cat.append(column)
        else:
            num.append(column)
    elif pd.api.types.is_datetime64_any_dtype(df[column]):
        date_time.append(column)
    else:
        cat.append(column)

print("Categorical columns:", cat)
print("Numerical columns:", num)
print("Date/Time columns:", date_time)
print("Identifier columns:", identifier)
```

## UNIVARIATE ANALYSIS
This notebook performs univariate analysis on various technology metrics for China and Japan. The metrics analyzed include:

- R&D Investment (in USD)
- Number of Patents Filed (Annual)
- Tech Exports (in USD)
- Internet Penetration (%)
- 5G Network Coverage (%)

The analysis includes visualizations such as box plots, histograms, and KDE plots to compare the distribution, central tendency, and variability of these metrics for both countries.


##### Filter data for China and Japan

```python

china_data = df[df['Country'] == 'China']
japan_data = df[df['Country'] == 'Japan']
```

##### Ploting Pie charts for visualizing Tech Sector in China and Japan

```python
china_tech_sectors = china_data.groupby('Tech Sector')['Market Share (%)'].sum()
japan_tech_sectors = japan_data.groupby('Tech Sector')['Market Share (%)'].sum()

colors = plt.get_cmap('tab20').colors

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
china_tech_sectors.plot.pie(autopct='%1.1f%%', startangle=90, colors=colors, shadow=True)
plt.title('China Tech Sectors Market Share')
plt.ylabel('')


plt.subplot(1, 2, 2)
japan_tech_sectors.plot.pie(autopct='%1.1f%%', startangle=90, colors=colors, shadow=True)
plt.title('Japan Tech Sectors Market Share')
plt.ylabel('')


plt.tight_layout()
plt.show()
```

```python

china_data['Country'] = 'China'
japan_data['Country'] = 'Japan'
combined_data = pd.concat([china_data, japan_data])


sns.set(style="whitegrid", palette="muted", rc={"axes.facecolor": ".9"})

metrics = [
    'Tech Exports (in USD)',
    'Internet Penetration (%)',
    '5G Network Coverage (%)',
    'Number of Patents Filed (Annual)',
    'Market Share (%)'
]


desc_stats = combined_data.groupby('Country')[metrics].describe()


print("Descriptive Statistics:")
print(desc_stats)


fig, axes = plt.subplots(len(metrics), 3, figsize=(27, 21))

for i, metric in enumerate(metrics):
    

    sns.boxplot(x='Country', y=metric, data=combined_data, ax=axes[i, 0], palette='Blues')
    axes[i, 0].set_title(f'Box Plot of {metric}')
    axes[i, 0].set_xlabel('Country')
    axes[i, 0].set_ylabel(metric)

    sns.histplot(data=combined_data, x=metric, hue='Country', kde=False, ax=axes[i, 1], palette='Set2', bins=20)
    axes[i, 1].set_title(f'Distribution of {metric}')
    axes[i, 1].set_xlabel(metric)
    axes[i, 1].set_ylabel('Frequency')
    

    sns.kdeplot(data=combined_data, x=metric, hue='Country', fill=True, ax=axes[i, 2], palette='Set2')
    axes[i, 2].set_title(f'KDE Plot of {metric}')
    axes[i, 2].set_xlabel(metric)
    axes[i, 2].set_ylabel('Density')

plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.suptitle("Univariate Analysis for Different Metrics by Country", fontsize=20)

plt.show()
```

## Multivariate Analysis

In this section, we explore relationships between multiple variables to identify patterns and correlations.

### Objectives

- Identify relationships between variables.
- Understand combined effects on outcomes.
- Visualize complex interactions.

### Techniques

1. **Heatmaps**: Visualize correlation matrices.
2. **Pair Plots**: Visualize pairwise relationships.
3. **Multivariate Regression**: Model relationships between dependent and independent variables.
4. **Principal Component Analysis (PCA)**: Reduce dimensionality and identify key variables.
5. **Cluster Analysis**: Group similar observations.


```python
correlation_cols = [
    'R&D Investment (in USD)', 
    'Number of Patents Filed (Annual)', 
    'Venture Capital Funding (in USD)', 
    'Number of Startups', 
    'Tech Exports (in USD)', 
    'Global Innovation Ranking'
]
```

#### Correlation Matrix 


```python

corr_matrix = combined_data[correlation_cols].corr()


plt.figure(figsize=(12, 8)) 
sns.set_theme(font_scale=1.2) 

heatmap = sns.heatmap(
    corr_matrix, 
    annot=True,               
    fmt='.2f',                
    cmap='coolwarm',          
    cbar_kws={'label': 'Correlation Coefficient'},  
    linewidths=0.5,           
    square=True              )

plt.title('Correlation Matrix of Key Metrics', fontsize=18, weight='bold', pad=20)
plt.xlabel('Metrics', fontsize=14, labelpad=10)
plt.ylabel('Metrics', fontsize=14, labelpad=10)


plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)


plt.tight_layout()
plt.show()
```

#### Scatter Plot between Market Share (%) and R&D Investment (in USD)

```python

plt.figure(figsize=(10, 6))  
sns.scatterplot(
    data=combined_data, 
    x='Market Share (%)', 
    y='R&D Investment (in USD)', 
    hue='Country', 
    style='Country',  
    palette='Set1',  
    s=100,            
    markers=["o", "s"], 
    edgecolor='black',  
    alpha=0.8         
)

plt.title('Market Share (%) vs. R&D Investment (in USD)', fontsize=16, weight='bold', pad=15)
plt.xlabel('Market Share (%)', fontsize=14)
plt.ylabel('R&D Investment (in USD)', fontsize=14)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()



```
